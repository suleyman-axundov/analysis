```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
```

```python
df_per_capital = pd.read_csv("co-emissions-per-capita (3).csv")
```

```python
df_per_capital.head()
```

```python
df_per_capital.info()
```

```python
df_per_capital_Aze = df_per_capital[df_per_capital['Entity'].str.contains('Azerbaijan')]
```

```python
df_per_capital_Aze_1990_2022 = df_per_capital_Aze[df_per_capital_Aze['Year']>=1990]
```

```python
df_per_capital_Aze_1990_2022
```

```python
df_per_capital_Aze_1990_2022['Percentage'] = df_per_capital_Aze['Annual CO₂ emissions (per capita)'].pct_change()* +100
```

```python
df_per_capital_Aze_1990_2022
```

```python
df_per_capital_Aze_1990_2022_group = df_per_capital_Aze_1990_2022.groupby(['Year'])['Annual CO₂ emissions (per capita)'].mean()
```

```python
df_per_capital_Aze_1990_2022_group
```

```python
df_per_capital_Aze_1990_2022_group
```

```python
plt.figure(figsize=(19,8))
plt.plot(df_per_capital_Aze_1990_2022_group,label = "Annual CO₂ emissions (per capita)")

# Highlight the highest and lowest points
max_value = df_per_capital_Aze_1990_2022_group.max()
min_value = df_per_capital_Aze_1990_2022_group.min()
max_index = df_per_capital_Aze_1990_2022_group.idxmax()
min_index = df_per_capital_Aze_1990_2022_group.idxmin()

# Mark the highest point
plt.plot(max_index, max_value, marker='^', color='green', markersize=8, label='Highest Value')
plt.text(max_index, max_value, f'{max_value:,.1f}', ha='center', va='bottom', color='green', fontsize=15)

# Mark the lowest point
plt.plot(min_index, min_value, marker='v', color='red', markersize=8, label='Lowest Value')
plt.text(min_index, min_value, f'{min_value:,.1f}', ha='center', va='top', color='red', fontsize=14)


plt.xlabel("Year",labelpad=10)
plt.ylabel("Annual CO₂ emissions (per capita)",labelpad=10)
plt.xticks(df_per_capital_Aze_1990_2022['Year'],rotation = 45)
plt.title("Annual CO₂ emissions (per capita) for Azerbaijan 1990-2022")
ax = plt.gca()
ax.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'{x:,.0f}t'))
plt.legend()
plt.grid()
plt.show()
```

```python
df_per_capital_Aze_1990_2022_group1 = df_per_capital_Aze_1990_2022.groupby(['Year'])['Percentage'].mean()
```

```python
df_per_capital_Aze_1990_2022_group1
```

```python
plt.figure(figsize=(19,8))
plt.plot(df_per_capital_Aze_1990_2022_group1,label = "Annual Percentage CO₂ emissions (per capita)")

# Highlight the highest and lowest points
max_value = df_per_capital_Aze_1990_2022_group1.max()
min_value = df_per_capital_Aze_1990_2022_group1.min()
max_index = df_per_capital_Aze_1990_2022_group1.idxmax()
min_index = df_per_capital_Aze_1990_2022_group1.idxmin()

# Mark the highest point
plt.plot(max_index, max_value, marker='^', color='green', markersize=8, label='Highest Value')
plt.text(max_index, max_value, f'{max_value:,.1f}%', ha='center', va='bottom', color='green', fontsize=15)

# Mark the lowest point
plt.plot(min_index, min_value, marker='v', color='red', markersize=8, label='Lowest Value')
plt.text(min_index, min_value, f'{min_value:,.1f}%', ha='center', va='top', color='red', fontsize=14)

plt.xlabel("Year",labelpad=10)
plt.ylabel("Annual CO₂ emissions (per capita)",labelpad=10)
plt.xticks(df_per_capital_Aze_1990_2022['Year'],rotation = 45)
plt.title("Annual Percentage CO₂ emissions (per capita) for Azerbaijan 1990-2022")
ax = plt.gca()
ax.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'{x:,.0f}%'))
plt.legend()
plt.grid()
plt.show()
```

```python
df_per_capital_Aze_2018_2022 = df_per_capital_Aze_1990_2022.tail(5)
```

```python
df_per_capital_Aze_2018_2022.index = [1,2,3,4,5]
```

```python
df_per_capital_Aze_2018_2022.drop(columns=["Code","Entity"],inplace=True)
```

```python
df_per_capital_Aze_2018_2022['Percentage'] = pd.to_numeric(df_per_capital_Aze_2018_2022['Percentage'], errors='coerce')

# Apply the formatting with ↑ and ↓ symbols and HTML coloring
df_per_capital_Aze_2018_2022['Percentage'] = df_per_capital_Aze_2018_2022['Percentage'].apply(lambda x: f'<span style="color:blue;">{x:,.1f}% ↑</span>' if x > 0 else f'<span style="color:red;">{x:,.1f}% ↓</span>')

# Display the DataFrame with HTML rendering (only works in a Jupyter notebook)
df_per_capital_Aze_2018_2022 = df_per_capital_Aze_2018_2022.style.set_caption("CO₂ emissions (per capita) Percentage Changes (2018-2022)").format({'Percentage': lambda x: x})
```
